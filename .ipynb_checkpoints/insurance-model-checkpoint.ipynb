```python
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.read_csv("./insurance.csv")
```

```python
df.head()
```

```python
# Convert Categorical Data into Boolean

# Annotations

# Male-1  Female-0

# Smoker(yes)-1  Smoker(no)-0 

# southwest -0
# southeast -1
# northwest -2
# northeast -3

df= df.replace({'sex':{'male':1,"female":0}})
df= df.replace({'smoker':{'yes':1,"no":0}})
df= df.replace({'region':{'southwest':0,"southeast":1,"northwest":2,"northeast":3}})

```

# Linear Regression using Scikit-learn

```python
pip install scikit-learn
```

```python
def rmse(targets, predictions):
    return np.sqrt(np.mean(np.square(targets - predictions)))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
inputs = df[['age','children','smoker','sex',"region"]]
targets = df['charges']

print('inputs.shape :', inputs.shape)
print('targes.shape :', targets.shape)
```

```python
model.fit(inputs, targets)
```

```python
predictions = model.predict(inputs)
```

```python
predictions
```

```python
rmse(targets, predictions)
```

```python
model.coef_
model.intercept_

```
